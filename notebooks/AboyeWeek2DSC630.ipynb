```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('data/Coursera.csv')
```

#### **Data Source: https://www.kaggle.com/datasets/siddharthm1698/coursera-course-dataset**

# **Data Cleaning**

```python
df.head()
```

```python
df.shape
```

### **Fill missing values with a placeholder (e.g., 'Unknown' for categorical data, 0 for numerical data)**

```python
df_cleaned = df.fillna({
    'Course Name': 'Unknown',
    'University': 'Unknown',
    'Difficulty Level': 'Unknown',
    'Course Rating': 0,
    'Course URL': 'Unknown',
    'Course Description': 'Unknown',
    'Skills': 'Unknown'
})
```

```python
# Removing duplicates
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df.shape
```

```python
df.info()
```

```python
# Standardizing text fields
```

```python
df_cleaned['Course Name'] = df_cleaned['Course Name'].str.lower().str.strip()
df_cleaned['University'] = df_cleaned['University'].str.lower().str.strip()
df_cleaned['Difficulty Level'] = df_cleaned['Difficulty Level'].str.lower().str.strip()
df_cleaned['Course URL'] = df_cleaned['Course URL'].str.lower().str.strip()
df_cleaned['Course Description'] = df_cleaned['Course Description'].str.lower().str.strip()
df_cleaned['Skills'] = df_cleaned['Skills'].str.lower().str.strip()
```

```python
df_cleaned.to_csv('data/cleaned_dataset.csv', index=False)
```

```python
# Display the first ten rows of the cleaned dataset
df_cleaned.head(10)
```

```python
# Display the last 10 rows of the cleaned dataset
df_cleaned.head(10)
```

```python
# Distribution of Course Ratings
plt.figure(figsize=(15, 8))
sns.histplot(df_cleaned['Course Rating'], bins=20, kde=True)
plt.title('Distribution of Course Ratings')
plt.xlabel('Course Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Courses by Difficulty Level
plt.figure(figsize=(10, 6))
sns.countplot(y='Difficulty Level', data=df_cleaned, order=df_cleaned['Difficulty Level'].value_counts().index)
plt.title('Courses by Difficulty Level')
plt.xlabel('Number of Courses')
plt.ylabel('Difficulty Level')
plt.show()
```

```python
# Top 10 Universities or Course Provider by Number of Courses
plt.figure(figsize=(10, 6))
top_universities = df_cleaned['University'].value_counts().head(10)
sns.barplot(x=top_universities.values, y=top_universities.index)
plt.title('Top 10 Universities by Number of Courses')
plt.xlabel('Number of Courses')
plt.ylabel('University')
plt.show()
```

```python
# Distribution Analysis of Course Ratings by Difficulty Level
plt.figure(figsize=(10, 6))
sns.violinplot(x='Difficulty Level', y='Course Rating', data=df_cleaned)
plt.title('Course Ratings by Difficulty Level')
plt.xlabel('Difficulty Level')
plt.ylabel('Course Rating')
plt.show()
```
